```python
import numpy as np


def get_median_filtered(signal, threshold=3):
    signal = signal.copy()
    difference = np.abs(signal - np.median(signal))
    median_difference = np.median(difference)
    if median_difference == 0:
        s = 0
    else:
        s = difference / float(median_difference)
    mask = s > threshold
    signal[mask] = np.median(signal)
    return signal
```

```python
figsize = (20, 5)
kw = dict(marker='o', linestyle='none', color='r', alpha=0.3)

df1['TempLand_medf'] = get_median_filtered(df1['TempLand'].values, threshold=3)

outlier_idx = np.where(df1['TempLand_medf'].values != df1['TempLand'].values)[0]

fig, ax = plt.subplots(figsize=figsize)
df1['TempLand'].plot()
df1['TempLand'][outlier_idx].plot(**kw)
_ = ax.set_ylim(-50, 50)
```

```python
figsize = (15,5)
kw = dict(marker='o', linestyle='none', color='r', alpha=0.3)

df1['TempGlobal_medf'] = get_median_filtered(df1['TempGlobal'].values, threshold=3)

outlier_idx = np.where(df1['TempGlobal_medf'].values != df1['TempGlobal'].values)[0]

fig, ax = plt.subplots(figsize=figsize)
df1['TempGlobal'].plot()
df1['TempGlobal'][outlier_idx].plot(**kw)
_ = ax.set_ylim(-50, 50)
```

```python
from sklearn.ensemble import IsolationForest
pd.set_option('display.max_rows',500)
pd.set_option('display.max_columns',500)
pd.set_option('display.width',1000)
data = df1 [['SeaLevel']]

outlierFraction = 4.0 / len (df1.index)
model = IsolationForest(contamination = outlierFraction)
model.fit(data)

df1['anomaly_sea'] = pd.Series (model.predict(data))
df1 ['anomaly_sea'] = df1 ['anomaly_sea'].map( {1: 0, -1: 1} )
#display(df1['anomaly_Coal'])
# Visualisation
fig, ax = plt.subplots ()
a = df1.loc [df1 ['anomaly_sea'] == 1, ['year', 'SeaLevel']]
ax.plot (list (df1 ['year']), list (df1 ['SeaLevel']), color = 'grey')
ax.scatter (list (a ['year']), list (a ['SeaLevel']), color = 'red', lw = 1)

plt.title ('Anomaly Detection: sealevel', fontweight = 'bold', fontsize = 'x-large')
plt.ylabel ('Values', fontweight = 'bold')
plt.ylabel ('year', fontweight = 'bold')
plt.xticks ()
plt.yticks ()
plt.show ()
```
