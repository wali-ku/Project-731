```python
# load csv into pandas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('C:\\Users\\Ronald\\Documents\\EECS_731\\Project\\global-fossil-fuel-consumption.csv')
df.fillna(method='ffill')
df.head()
```

```python
# clean up data to include only years 1900 and later
df = df[df.Year >= 1900]
df.head()
```

```python
# make 3 plots of year vs. coal consumption, year vs. crude oil consumption, and year vs. natural gas consumption
year = df['Year']
coal = df['Coal (terawatt-hours)']
oil = df['Crude oil (terawatt-hours)']
gas = df['Natural gas (terawatt-hours)']

plt.plot(year, coal, linewidth=3, label='Coal')
plt.title('Coal Consumption vs. Year')  
plt.xlabel('Year')  
plt.ylabel('Consumption (terrawatt-hours)')
plt.legend(loc='center left', bbox_to_anchor=(1.05,.7))
plt.show()
```

```python
plt.plot(year, oil, linewidth=3, label='Crude Oil')
plt.title('Crude Oil Consumption vs. Year')  
plt.xlabel('Year')  
plt.ylabel('Consumption (terrawatt-hours)')
plt.legend(loc='center left', bbox_to_anchor=(1.05,.7))
plt.show()
```

```python
plt.plot(year, gas, linewidth=3, label='Natural Gas')
plt.title('Natural Gas Consumption vs. Year')  
plt.xlabel('Year')  
plt.ylabel('Consumption (terrawatt-hours)')
plt.legend(loc='center left', bbox_to_anchor=(1.05,.7))
plt.show()
```

```python
# Place coal, oil, and gas consumption on same plot
plt.plot(year, coal, linewidth=3, label='Coal')
plt.plot(year, oil, linewidth=3, label='Crude Oil')
plt.plot(year, gas, linewidth=3, label ='Natural Gas')
plt.title('Overall Consumption vs. Year')  
plt.xlabel('Year')  
plt.ylabel('Consumption (terrawatt-hours)')
plt.legend(loc='center left', bbox_to_anchor=(1.05,.7))
plt.show()
```
