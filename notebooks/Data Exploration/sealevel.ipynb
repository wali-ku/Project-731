```python
from sklearn.linear_model import LinearRegression
import numpy as np
import pandas as pd
DIR='/Users/ravali/Desktop'
df1 = pd.read_csv(DIR+'/sealevel.csv', delimiter=',')
#display(df1)
#display(df1.dtypes)
df1['year'] = pd.DatetimeIndex(df1['Time']).year
#display(df1.year)
#df1.drop(df1.Time,axis=1)
df1.drop('Time', axis=1, inplace=True)
df1=df1.dropna()
display(df1)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.pointplot(x=df1['year'], y=df1['GMSL'],data=df1,color='lime',alpha=0.8)
plt.xticks(rotation= 90,ha="right")
plt.xlabel('year')
plt.ylabel('GMSL')
plt.rcParams["xtick.labelsize"] = 7
plt.title('rise of sealevel from 1',fontsize=20)
plt.show()
```

```python
plt.rcParams['agg.path.chunksize'] = 10000
df1.plot (x='year',y='GMSL',figsize = (15, 5))
plt.xlabel ('Year', fontweight = 'bold', fontsize = 'x-large')
plt.ylabel ('GMSL', fontweight = 'bold', fontsize = 'x-large')
plt.xticks (fontweight = 'bold',fontsize=10)
plt.yticks (fontweight = 'bold',fontsize=10)
plt.title('Sea level rise over years',fontsize=20)
plt.show ()
#sns.pointplot(x=df1['year'], y=df1['GMSL'],data=df1,color='lime',alpha=0.8)
```
