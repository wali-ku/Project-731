```python
%%time

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from  datetime import datetime as dt
plt.style.use('fivethirtyeight')
%matplotlib inline
```

## Preliminary Setup - Loading and Cleaning the Dataset
---

```python
df_pop = pd.read_csv ('../data/firstPass/population.csv', index_col = ['Year'])
df_pop.head ()
```

## Observation
---
Dataframe is in reverse chronological order. We should fix that.

```python
df_popSorted = df_pop.reindex (index = df_pop.index [::-1])
df_popSorted.head ()
```

```python
df_popSorted.plot (figsize = (15, 5))
plt.xlabel ('Year', fontweight = 'bold')
plt.ylabel ('Population', fontweight = 'bold')
plt.xticks (fontweight = 'bold')
plt.yticks (fontweight = 'bold')
plt.title ('Yearly Global Population', fontweight = 'bold', fontsize = 'x-large')
plt.show ()
```
